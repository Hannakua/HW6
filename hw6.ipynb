```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_mnist = pd.read_csv("data/mnist.csv", header=None)
data_mnist.head(7)
```

```python
data_mnist.shape
```

```python
data_2d = pd.read_csv("data/data_2d.csv", header=None)
data_2d.head(15)
```

```python
data_2d.shape
```

```python
X = data_mnist.to_numpy()[:, 0:]
y = data_mnist.to_numpy()[:, 0]
```

```python
from sklearn.cluster import KMeans
inertions_mnist = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, verbose=0)
    model.fit(X)
    inertions_mnist.append(model.inertia_)
# print(len(inertions))
```

```python
plt.plot(list(range(1,11)), inertions_mnist)
plt.show()
```

```python
# по графіку можна припустити, що оптимальна кількість кластерів для датасету data_mnist дорівнює трьом.
```

```python
inertions_2d = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, verbose=0)
    model.fit(data_2d)
    inertions_2d.append(model.inertia_)
```

```python
plt.plot(list(range(1,11)), inertions_2d)
plt.show()
```

```python
# по графіку можна побачити, що оптимальна кількість кластерів для датасету data_2d дорівнює двом.
```

```python
kmeans = KMeans(n_clusters=2)
X = data_2d.iloc[:, 1]
y = data_2d.iloc[:, 2]
c = kmeans.fit_predict(data_2d)
plt.scatter(X, y, c=c)
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
kmeans = KMeans(n_clusters=3)
c = kmeans.fit_predict(data_mnist)
```

```python
X_reduced_mnist = PCA(n_components=2).fit_transform(data_mnist)

plt.scatter(X_reduced_mnist[:, 0], X_reduced_mnist[:, 1], c=c)
plt.xlabel("1st Eigenvector")
plt.ylabel("2nd Eigenvector")
plt.show()
```
